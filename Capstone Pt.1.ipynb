## Predicting Health Insurance Status Using the 2023 MEPS: A Machine-Learning and Vulnerability-Index Approach to Understanding Access Gaps
Aiyana Wright

### Abstract
Persistent disparities in health insurance coverage in the United States continue to shape access to care, affordability, and long-term health outcomes. This study uses 2023 Medical Expenditure Panel Survey (MEPS) microdata to develop a predictive modeling framework that classifies individuals into private insurance, public-only insurance, or uninsured categories. In addition, it introduces a novel PCA-based Vulnerability Index designed to capture socioeconomic and health-utilization disadvantage. Using a combination of multinomial logistic regression, Random Forests, Gradient Boosting, and XGBoost, the study evaluates predictors of insurance status and assesses model performance using accuracy, F1 scores, and SHAP interpretability. XGBoost achieved the strongest performance (75% accuracy), with income, education, employment status, and healthcare utilization emerging as the most influential predictors. Higher vulnerability scores were strongly associated with an increased likelihood of lacking private insurance. Findings highlight enduring racial and socioeconomic disparities, demonstrating the utility of machine learning and dimensionality-reduction techniques in informing policy and targeted interventions.

### Introduction
Health insurance coverage is a foundational determinant of healthcare access in the United States, yet disparities remain substantial across socioeconomic, racial, and demographic groups. Even after the Affordable Care Act expanded coverage, gaps persist in who holds private insurance, who relies solely on public programs, and who remains uninsured. These disparities affect individuals' ability to access preventive care, manage chronic conditions, and avoid incurring burdensome out-of-pocket costs.
Understanding which factors most strongly predict insurance statusand how structural vulnerability contributes to uninsuranceremains crucial for informed public policy, effective outreach efforts, and risk-stratification models. This study utilizes nationally representative 2023 MEPS Household Component data to explore these questions and develop a predictive model that identifies populations at heightened risk of lacking private insurance.
#### Three research questions guide this work:
1.	Which sociodemographic and health-use factors predict an individuals insurance status?
2.	Can a PCA-derived vulnerability score effectively summarize structural disadvantage and improve predictive performance?
3.	Which populations are most vulnerable to lacking private insurance, and what drives these disparities?
This article presents the methodological framework, preliminary findings, and implications for future research and policy action.
### Data and Measures
#### Data Source
The analysis utilizes the 2023 Medical Expenditure Panel Survey (MEPS) Household Component, a nationally representative dataset that contains detailed demographic, socioeconomic, health status, utilization, and insurance measures.
	Sample size: 18,919 individuals

	Time period: 2023 full-year consolidated file
Target variable: INSURANCE_TYPE_CLEANED
- 	0 = Private
- 	1 = Public-only
-	2 = Uninsured
### Key Predictive Features
	Demographics: age, race/ethnicity, education years, family size

	Socioeconomic indicators: family income, employment status

	Healthcare utilization: dental visits, office visits, emergency room visits

	Composite index: PCA-derived Vulnerability Score (PC1)

	Health ratings: mental health, overall health
### Data Preparation
Data cleaning followed MEPS recommendations and standard preprocessing for machine learning:
	Removed individuals with missing insurance information

	Standardized continuous predictors

	Recoded categorical education categories into numeric years

	Winsorized extreme outliers (top/bottom 0.5%)

	Constructed PCA Vulnerability Index from socioeconomic and utilization indicators

The final dataset was exported for modeling and dashboard visualization.
### Literature Review
Prior machine-learning studies in healthcare have demonstrated that ensemble methods, such as Random Forests and XGBoost, often outperform traditional models when predicting complex outcomes. Rahman and Dua (2025), for example, used Louisiana Medicaid claims to predict clinical utilization, COVID-19 infection, and behavioral risk, finding that XGBoost models reached accuracies exceeding 95%, substantially outperforming logistic regression.
#### Their work highlights several themes relevant to insurance prediction:
	Nonlinear interactionscommon in socioeconomic determinantsare best captured by tree-based models.

	Age, disability, chronic illness, and region consistently rank as powerful predictors of health behaviors and outcomes.

	Ensemble models manage high-dimensional data more effectively while balancing accuracy and interpretability.

### Methodology
#### Modeling Approach
A supervised multiclass classification strategy was used to predict insurance status. Four model families were trained and compared:
1.	ANOVA + Chi- Squared Test 
2.	PCA Vulnerability Index 
3.	Multinomial Logistic Regression
4.	Random Forest
5.	Gradient Boosting
6.	XGBoost (final selection)
XGBoost was chosen for its ability to model nonlinear relationships, handle mixed data types, and produce interpretable feature-attribution outputs.
### Technical Stack
	Python: Pandas, NumPy, Scikit-learn, XGBoost, SHAP

	Tableau: heatmaps, scatterplots, dashboard elements

	Jupyter Notebook: exploratory analysis and model training
### Evaluation Strategy
	Train/Test split: 80/20

	Cross-validation: 5-fold

	Metrics: Accuracy, macro F1, confusion matrix

	Interpretability: SHAP feature importance

### Discussion
### Interpretation of Findings
The modeling pipeline successfully answers the research questions:
	Predictors of insurance status: Income, education, and employment are dominant drivers.

	Role of vulnerability: Higher PCA-based vulnerability scores significantly increase the probability of public-only insurance or uninsurance.

	High-risk populations: Racial/ethnic disparities appear consistently, particularly among Black and Hispanic adults with lower incomes.
### Comparison to Existing Research
Findings align with prior studies, which show accuracy ranges of 7080% for insurance prediction models, and reaffirm socioeconomic determinants as the core drivers. This project expands existing work by incorporating PCA and SHAP explainability tools.
### Limitations
	Imputed education may introduce measurement error.

	Public-only vs. uninsured remains challenging to distinguish.

	MEPS self-reported nature may introduce recall bias.

	PCA reduces complex disadvantage into a single dimension, which may obscure nuance.
### Future Directions
	Test additional models such as CatBoost and LightGBM

	Incorporate chronic-condition severity and unmet-need measures into the vulnerability index

	Expand dashboard with group-specific accuracy (e.g., race  income)

	Evaluate fairness metrics such as demographic parity and equal opportunity

### Conclusion
This study demonstrates the value of combining machine learning, dimensionality reduction, and explainability methods to understand U.S. health insurance disparities. The results show that structural vulnerability, captured through a PCA-based index, strongly predicts insurance status and aligns with known socioeconomic inequalities. As policymakers and health systems seek to improve outreach, risk scoring, and targeted enrollment strategies, these methods offer a scalable and interpretable toolkit grounded in nationally representative data.
